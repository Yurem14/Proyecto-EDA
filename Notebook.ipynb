```python
import pandas as pd
import os
import random
import time
```

```python
summer = pd.read_csv('Data/Athletes_summer_games.csv')
winter = pd.read_csv('Data/Athletes_winter_games.csv')
regions = pd.read_csv('Data/regions.csv')
df = pd.concat([summer,winter])
del summer, winter
```

```python
nulos = {'Age': int(df['Age'].isnull().sum())}
nulos['Year']=int(df['Year'].isnull().sum())
nulos['Name']=int(df['Name'].isnull().sum())
nulos['Sex']=int(df['Sex'].isnull().sum())
nulos['Team']=int(df['Team'].isnull().sum())
nulos['NOC']=int(df['NOC'].isnull().sum())
nulos['Games']=int(df['Games'].isnull().sum())
nulos['Season']=int(df['Season'].isnull().sum())
nulos['City']=int(df['City'].isnull().sum())
nulos['Sport']=int(df['Sport'].isnull().sum())
nulos['Event']=int(df['Event'].isnull().sum())
nulos['Medal']=int(df['Medal'].isnull().sum())
```

```python
import matplotlib.pyplot as plt
# Crear la gráfica de barras
plt.bar(nulos.keys(), nulos.values())
# Añadir títulos y etiquetas
plt.title('Nulos')
plt.xlabel('Categorías')
plt.ylabel('Valores')

# Mostrar la gráfica
plt.show()
```

```python
filtro=df[df['Medal']=='Gold']
filtro=df[df['Medal'].notnull()]
data = filtro['Team'].value_counts().head(10)
# Crear la gráfica de barras
plt.bar(data.keys(), data.values)
# Añadir títulos y etiquetas
plt.title('Medallas de ORO')
plt.xlabel('Paises')
plt.xticks(rotation=90)
plt.ylabel('Cantidad')

# Mostrar la gráfica
plt.show()
```
